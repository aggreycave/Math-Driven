### Project: Math-Driven Linear Regression.

```python
# 1. Applying Linear Algebra, Calculus, and Statistics.

    # Linear Algebra.
        # Mathematical Model:
            # y = XQ.
                # X: Feature matrix (with bias term).
                # Q: Parameters (wights and bias).
                # y^: Predicted values.
    # Calculus.
        # Optimisation of Q involves minimizing the loss function.
            # J(Q) = 1/2m(mMi=1(y^i - yi)^2).
            # VJ = 1/mX^T(XQ - y).
    # Statistics.
        # Metrics like Mean Squared Error (MSE) and R^2 are used to evaluate model performance

# 2. Using Gradient Descent for Parameter Optimization.
    # Gradient Descent Algorithem.
        # Iteratively update Q using Q := Q - aVJ
            # a: Learning rate.
    # Key Steps: 
        # Initilise parameters (Q).
        # Compute gradients (Vj).
        # Update parameters using the gradeint descent rule.

# 3. Evaluating the Model Using Statistical Metrics.
    # Mean Square Error (MSE).
        # MSE = 1/m(mEi=1(y^i - yi)^2).
            # Measures the average squared error.
    # R-squared (R^2).
        # R^2 = 1 - SSres/SStot.
            # Measures how well the regression line explians the variance un the data.
```

##### Hands on Project: Linear Regression from Scratch.

```python
import numpy as np

# Generate Synthetic data.
np.random.seed(42)
X = np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Add bias term to the feature matrix.
X_b = np.c_[np.ones((100, 1)), X]

# Initialising the parameters.
theta = np.random.randn(2, 1)
learning_rate = 0.1
iterations = 1000

# Task 1: Implement the Mathematical Formula for Linear Regression.

# Creating a predict model.
def predict(X, theta):
    return np.dot(X, theta)
    
# Task 2: Use Gradient Descent to Optimise the Model Parameters.

def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    for _ in range(iterations):
        gradients = (1/m) * np.dot(X.T, (np.dot(X, theta) - y))
        theta -= learning_rate * gradients
    return theta    
    
# Task 3: Calculute Evaluation Metrices.

def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def r_squared(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred) **2 )
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - (ss_res / ss_tot)

# Perform gradient descent.
theta_optimized = gradient_descent(X_b, y, theta, learning_rate, iterations)

# Predictions and Evaluations.
y_pred = predict(X_b, theta_optimized)
mse = mean_squared_error(y, y_pred)
r2 = r_squared(y, y_pred)

print("Optimisied Parameters (Theta): ", theta_optimized) # Parameters are good if between 4 and 3
print("Mean Squared Error (MSE): ", mse)
print("R Squared  (R2): ", r2)
```
